```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Load Data

df = pd.read_csv('/content/student_performance_dataset.csv')
 # Replace with your actual CSV file path
print("Original Data Shape:", df.shape)
print(df.head())
```

```python
# 2. Basic Info

print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# 3. Handle Missing Values

df['Parental_Education_Level'] = df['Parental_Education_Level'].fillna('')
df = df.drop_duplicates()
print("\nAfter Removing Duplicates:", df.shape)
```

```python
# 4. Encode Categorical Variables

# Encodes Pass_Fail to numbers
le = LabelEncoder()
df['Pass_Fail'] = le.fit_transform(df['Pass_Fail'])
print("\nPass_Fail Encoding Done!")
print(df['Pass_Fail'].value_counts())

# Convert Parental_Education_Level text to numbers using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=3000)
X = vectorizer.fit_transform(df['Parental_Education_Level'])
y = df['Pass_Fail']

print("\nTF-IDF Matrix Shape:", X.shape)
```

```python
num_cols = ['Study_Hours_per_Week', 'Attendance_Rate', 'Past_Exam_Scores', 'Final_Exam_Score']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
```

```python
# 6. Feature Scaling

# No StandardScaler needed - TF-IDF already scales the text data
```

```python
target_col = 'Pass_Fail'

if target_col in df.columns:
    X = vectorizer.fit_transform(df['Parental_Education_Level'])
    y = df[target_col]

    print("\nTarget Column:", target_col)
    print("Label Distribution:\n", df[target_col].value_counts())
```

```python
import numpy as np
import sklearn.preprocessing as preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = vectorizer.fit_transform(df['Parental_Education_Level'])
y = np.asarray(df['Pass_Fail'])

# No StandardScaler needed - TF-IDF already scales the text data

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size = 0.3, random_state = 4)

print ('Train set:', X_train.shape, y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

```python
#1.Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#2.K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#3.Decision Tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#4.Random Forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#5.Support Vector Machine
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#6.Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#7.Naive Bayes
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```
